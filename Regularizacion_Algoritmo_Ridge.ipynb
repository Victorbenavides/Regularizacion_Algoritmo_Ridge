**Regularización Algoritmo Ridge**

```python
#Importamos librerias
from sklearn.datasets import make_regression
from matplotlib import pyplot as plt
import numpy as np
from sklearn.linear_model import Ridge
```

```python
#Generamos los datos de muestra
X, y, coefficients = make_regression(
    n_samples=50,
    n_features=1,
    n_informative=1,
    n_targets=1,
    noise=5,
    coef=True,
    random_state=1
)
```

```python
#Definimos el hiperparámetro alfa. Alfa determina la fuerza de regularización
alpha = 1
```

```python
#Creamos la matriz de identidad en base a la muestra
n, m = X.shape
I = np.identity(m)
w = np.dot(np.dot(np.linalg.inv(np.dot(X.T, X) + alpha * I), X.T), y)
w
```

```python
#Imprimimos coeficientes utilizados para generar datos y se observa que la matriz no esta muy lejos de los coeficientes (correlación muy alta)
coefficients
```

```python
#Graficamos y observamos la linea de regresion
plt.scatter(X, y)
plt.plot(X, w*X, c='red')
```

```python
#Repetimos el mismo proceso pero ahora aplicando algoritmo ridge (En este caso alpha al ser 1 no hay cambios significativos)
rr = Ridge(alpha=1)
rr.fit(X, y)
w = rr.coef_
```

```python
w
```

```python
plt.scatter(X, y)
plt.plot(X, w*X, c='red')
```

```python
#Aqui se logra ver el efecto que los hiperparametros de alpha en ridge al ir aumentando la regularizacion se va notando mas
rr = Ridge(alpha=10)
rr.fit(X, y)
w = rr.coef_[0]
plt.scatter(X, y)
plt.plot(X, w*X, c='red')
```

```python
#Y aqui se ve que alpha al 100, la línea de regresión ya no es un ajuste perfecto
rr = Ridge(alpha=100)
rr.fit(X, y)
w = rr.coef_[0]
plt.scatter(X, y)
plt.plot(X, w*X, c='red')
```
